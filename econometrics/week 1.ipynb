```python
import pandas as pd
import seaborn as sns
import numpy as np
sns.set(color_codes=True)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Dataset1.xls")
```

```python
df
```

```python
# x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 12
n, bins, patches = plt.hist(df["Sales"], num_bins, facecolor='blue', alpha=0.5)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# The sale manager can decide the price level as :
num_bins = 7
n, bins, patches = plt.hist(df["Price"], num_bins, facecolor='blue', alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Frequency') #  # of weeks
plt.show()
```

```python
# Create data
# N = 500
x = df["Price"]
y = df["Sales"]
# color = (0,0,0)
area = np.pi*3

# Plot
plt.scatter(x, y, s=area, alpha=0.5)
plt.title('Scatter Diagram of Sales aganist Price')
plt.xlabel('PRICE')
plt.ylabel('SALES')
plt.show()
```

```python
# Create data
# N = 500
x = df["Price"]
y = df['Sales']/df['Sales'].sum()
# color = (0,0,0)
area = np.pi*3

# Plot
plt.scatter(x, y, s=area, alpha=0.5)
plt.title('Scatter Diagram of Average Sales aganist Price')
plt.xlabel('PRICE')
plt.ylabel('AVERAGE SALES')
plt.show()
```
